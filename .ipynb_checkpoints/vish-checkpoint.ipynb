```python
import pandas as pd
import numpy as np 
import tensorflow as tf
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize
```

```python
train = pd.read_excel('training_set_rel3.xlsx')
print(train.shape)
train['length'] = train['essay'].str.split().str.len()
train['avg_word_len'] = train['essay'].str.split().apply(lambda x: np.mean([len(y) for y in x]))
train['avg_sent_len'] = train['essay'].str.split().apply(lambda x: np.mean([len(y) for y in sent_tokenize(x)]))

train['std_word_len'] = train['essay'].str.split().apply(lambda x: np.std([len(y) for y in x]))
train['unique_words'] = train['essay'].str.split().apply(lambda x: len(np.unique(x)))
train
```

```python
c = train.groupby('essay_set').mean().length
a = train.groupby('essay_set').count().essay
b = train['essay_set'].unique()
plt.bar(b,a)
plt.ylabel('Number of Examples')
plt.xlabel('Prompt')
plt.figure()
plt.bar(b,c)
plt.ylabel('Length')
plt.xlabel('Prompt')
```

```python
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)

for i in range(1,9):
    essay_set1 = train[train['essay_set'] == i]
    plt.scatter(essay_set1.length, essay_set1.domain1_score)
    plt.ylabel('Domain1_score of Essay Set ' + str(i))
    plt.xlabel('Length')
    plt.figure()
```

```python
figure(figsize=(8, 6), dpi=80)

for i in range(1,9):
    essay_set1 = train[train['essay_set'] == i]
    plt.scatter(essay_set1.avg_word_len, essay_set1.domain1_score)
    plt.ylabel('Domain1_score of Essay Set ' + str(i))
    plt.xlabel('Average Word Length')
    plt.figure()
```

```python
figure(figsize=(8, 6), dpi=80)

for i in range(1,9):
    essay_set1 = train[train['essay_set'] == i]
    plt.scatter(essay_set1.unique_words, essay_set1.domain1_score)
    plt.ylabel('Domain1_score of Essay Set ' + str(i))
    plt.xlabel('Unique Words')
    plt.figure()
```
